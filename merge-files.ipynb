Preprocessing for Final Project

```python
# Merging applications.csv and application_genres.csv
def merge_and_aggregate_data(genre_filepath, main_filepath, output_filepath):
    """
    Loads two CSV files, aggregates duplicate genre IDs, and merges the result.
    """
    
    # 2.1 Load the datasets
    try:
        df_genre = pd.read_csv(genre_filepath)
        df_main = pd.read_csv(main_filepath)
    except FileNotFoundError as e:
        print(f"Error: One of the input files was not found: {e.filename}")
        return
    except Exception as e:
        print(f"An error occurred while reading the CSV files: {e}")
        return

    print("Step 1: Data loaded successfully.")
    print(f"Main Data Rows: {len(df_main)}, Genre Rows: {len(df_genre)}")

    # 2.2 Aggregate the genre_ids for each unique appid
    # We group by 'appid' and aggregate all 'genre_id' values into a single list.
    df_aggregated_genre = df_genre.groupby('appid')['genre_id'].apply(list).reset_index()
    
    # Rename the aggregated column to make it clear
    df_aggregated_genre.columns = ['appid', 'genre_id_list']

    print("Step 2: Genre IDs aggregated into lists for each unique App ID.")
    # Optional: Convert the list of IDs into a comma-separated string for easier viewing in CSV
    df_aggregated_genre['genre_id_csv'] = df_aggregated_genre['genre_id_list'].apply(lambda x: ', '.join(map(str, x)))
    
    # We will use the list column for merging but save both in the output.
    df_aggregated_genre.drop(columns=['genre_id_list'], inplace=True)


    # 2.3 Merge the aggregated genres with the main data
    # We use a 'left' merge to keep all rows from the main data file (df_main).
    # If an appid exists in df_main but not in the genre file (like appid 104 in the mock data), 
    # the genre columns will be filled with NaN.
    df_merged = pd.merge(
        df_main, 
        df_aggregated_genre, 
        df_genre,
        on='appid', 
        how='left'
    )

    print("Step 3: Data merged based on 'appid'.")
    print(f"Final Merged Rows: {len(df_merged)}")

    # 2.4 Save the result to a new CSV file
    df_merged.to_csv(output_filepath, index=False)
    
    print(f"\nSuccess! Merged data saved to: {output_filepath}")
    print("\n--- Example of Merged Data (First 5 Rows) ---")
    print(df_merged.head())

```

```python
# Run the merge process
merge_and_aggregate_data('application_genres.csv', 'applications.csv', 'merged_app_data.csv')
```
